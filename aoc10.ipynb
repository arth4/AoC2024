```python
scores = [[None for y in range(inp.shape[1])] for x in range(inp.shape[0])]
```

```python
def legal_moves(x, y):
    return [
        (x + dx, y + dy)
        for dx, dy in ((0, 1), (0, -1), (1, 0), (-1, 0))
        if (0 <= x + dx < inp.shape[0])
        and (0 <= y + dy < inp.shape[1])
        and inp[x + dx, y + dy] == inp[x, y] + 1
    ]


def explore(x, y):
    if not scores[x][y] is None:
        return scores[x][y]

    if inp[x, y] == 9:
        scores[x][y] = {(x, y)}
    else:
        scores[x][y] = set().union(*[explore(xx, yy) for xx, yy in legal_moves(x, y)])

    return scores[x][y]
```

```python
sum(len(explore(x, y)) for x,y in np.argwhere(inp==0))
```

```python
def explore(x, y):
    if not scores[x][y] is None:
        return scores[x][y]

    if inp[x, y] == 9:
        scores[x][y] = 1
    else:
        scores[x][y] = sum([explore(xx, yy) for xx, yy in legal_moves(x, y)])

    return scores[x][y]
```
