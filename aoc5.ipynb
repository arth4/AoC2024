```python
def jagged2masked(ragged_list, fill_value=999999):
    max_len = max(len(sublist) for sublist in ragged_list)

    padded_array = np.array(
        [sublist + [np.nan] * (max_len - len(sublist)) for sublist in ragged_list],
        dtype=float  # Use float to support np.nan
    )

    mask = np.isnan(padded_array)

    if fill_value is not None:
        padded_array[mask] = fill_value

    masked_array = np.ma.masked_array(padded_array, mask=mask)

    return masked_array
```
